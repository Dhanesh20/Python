```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
df=pd.read_csv(r"C:\Users\prashant\Downloads\GroceryStoreDataSet.csv",header=None,names=['Product'])
```

```python
df
```

```python
def breakdata(data):
    return data.split(",")
```

```python
df['Product'].apply(breakdata)
```

```python
data=list(df['Product'].apply(breakdata))
```

```python
data
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te=TransactionEncoder()
finaldata=te.fit_transform(data)
```

```python
finaldata
```

```python
print(te.columns_)
```

```python
df=pd.DataFrame(finaldata,columns=te.columns_)
df.replace(True,1,inplace=True)
df.replace(False,0,inplace=True)
```

```python
df
```

```python
scores=apriori(df,min_support=0.2,use_colnames=True)
```

```python
scores
```

```python
score_rule=association_rules(scores,min_threshold=0.5)
```
