Write a program to show how Numpy taking less memory compared to Python List?

```python

import sys

# Create a NumPy array with 1000 integers
numpy_array = np.arange(1000)

# Create a Python list with the same 1000 integers
python_list = list(range(1000))

# Calculate memory usage for NumPy array and Python list
numpy_memory = numpy_array.nbytes
python_memory = sys.getsizeof(python_list)

# Print the memory usage
print("Memory usage of NumPy array:", numpy_memory, "bytes")
print("Memory usage of Python list:", python_memory, "bytes")

# Compare memory usage
if numpy_memory < python_memory:
    print("NumPy array uses less memory than Python list.")
elif numpy_memory > python_memory:
    print("Python list uses less memory than NumPy array.")
else:
    print("Both have the same memory usage.")
```

```python
df=pd.read_csv(r"C:\Users\prashant\Downloads\cars.csv")
df
```

```python

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
from scipy import stats


# Step 2: Find the datatype of all columns
data_types = df.dtypes

# Step 3: Find the number of missing values in each column
missing_values = df.isnull().sum()



numeric_columns = df.select_dtypes(include=['int64', 'float64'])
imputer = SimpleImputer(strategy='mean')
df[numeric_columns.columns] = imputer.fit_transform(numeric_columns)

data_description = df.describe()


z_scores = stats.zscore(df[numeric_columns.columns])
df = df[(z_scores < 3).all(axis=1)]


categorical_columns = df.select_dtypes(include=['object'])
label_encoder = LabelEncoder()
df[categorical_columns.columns] = df[categorical_columns.columns].apply(label_encoder.fit_transform)


print("Step 1: Data Types")
print(data_types)
print("\nStep 2: Missing Values")
print(missing_values)
print("\nStep 5: Filled Missing Values")
print(df)
print("\nStep 6: Data Description")
print(data_description)
print("\nStep 7: Data without Outliers")
print(df)
```
